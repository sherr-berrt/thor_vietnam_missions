```python
# 1965, 1966
vt_65 = pd.read_csv('Data/VietNam_1965.csv', low_memory = False)
vt_661 = pd.read_csv('Data/VietNam_1966_0.csv', low_memory = False)
vt_662 = pd.read_csv('Data/VietNam_1966_1.csv', low_memory = False)
```

```python
# 1967
vt_671 = pd.read_csv('Data/VietNam_1967_0.csv', low_memory = False)
vt_672 = pd.read_csv('Data/VietNam_1967_1.csv', low_memory = False)
vt_673 = pd.read_csv('Data/VietNam_1967_2.csv', low_memory = False)
vt_674 = pd.read_csv('Data/VietNam_1967_Oct.csv', low_memory = False)
```

```python
# 1968
vt_681 = pd.read_csv('Data/VietNam_1968_0.csv', low_memory = False)
vt_682 = pd.read_csv('Data/VietNam_1968_1.csv', low_memory = False)
vt_683 = pd.read_csv('Data/VietNam_1968_2.csv', low_memory = False)
vt_684 = pd.read_csv('Data/VietNam_1968_3.csv', low_memory = False)
```

```python
# Missions by US targeting NViet
vt_65['MSNDATE'] = vt_65['MSNDATE'].str.replace('-', '')
filt_65 = vt_65[(vt_65.TGTCOUNTRY == 'NORTH VIETNAM') & (vt_65.COUNTRYFLYINGMISSION == 'UNITED STATES OF AMERICA')]
```

```python
# Combine 1966 DFs together
vt_66 =(vt_661.append(vt_662))
vt_66['MSNDATE'] = vt_66['MSNDATE'].str.replace('-', '')

# Missions by US targeting NViet
filt_66 = vt_66[(vt_66.TGTCOUNTRY == 'NORTH VIETNAM') & (vt_66.COUNTRYFLYINGMISSION == 'UNITED STATES OF AMERICA')]
```

```python
# Combine 1967 DFs together
vt_67 =vt_671.append(vt_672).append(vt_673)
vt_67['MSNDATE'] = vt_67['MSNDATE'].str.replace('-', '')

# Fix dates from October
lamdafunc = lambda x: "196710" + x[2:].replace(x[-2:],'') if len(x) == 6 else "1967100" + x[2:].replace(x[-2:],'')
vt_674['MSNDATE'] = (vt_674['MSNDATE'].str.replace('/', '')).apply(lamdafunc)
vt_67 = vt_67.append(vt_674)
# Missions by US targeting NViet
filt_67 = vt_67[(vt_67.TGTCOUNTRY == 'NORTH VIETNAM') & (vt_67.COUNTRYFLYINGMISSION == 'UNITED STATES OF AMERICA')]
```

```python
# Combine 1968 DFs together
vt_68 =vt_681.append(vt_682).append(vt_683).append(vt_683)
vt_68['MSNDATE'] = vt_68['MSNDATE'].str.replace('-', '')
filt_68 = vt_68[(vt_68.TGTCOUNTRY == 'NORTH VIETNAM') & (vt_68.COUNTRYFLYINGMISSION == 'UNITED STATES OF AMERICA')]
```
